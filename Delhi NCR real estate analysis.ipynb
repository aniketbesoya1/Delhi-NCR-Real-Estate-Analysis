```python
mydata['Balcony'] = pd.to_numeric(mydata['Balcony'], errors='coerce')
```

```python
mydata['Status'] = mydata['Status'].fillna(mydata['Status'].mode()[0])
```

```python
mydata.isnull().sum()
```

```python
mydata['parking']=mydata['parking'].fillna(mydata['parking'].median())
```

```python
mydata.isnull().sum()
```

```python
mydata['Furnished_status'] = mydata['Furnished_status'].fillna(mydata['Furnished_status'].mode()[0])
```

```python
mydata.isnull().sum()
```

```python
mydata['Landmarks'] = mydata['Landmarks'].fillna('Not Mentioned')
```

```python
mydata['Lift']=mydata['Lift'].fillna(mydata['Lift'].median())
```

```python
cols= ['parking','Lift'] 
mydata[cols] = mydata[cols].round().astype(int)
```

 # highlight the columns

```python
mydata.columns = mydata.columns.str.upper()
```

```python
mydata.head()
```

# Removing unwanted columns

```python
mydata = mydata.drop(['DESC'],axis=1)
```

```python
mydata.head()
```

```python
mydata.columns =mydata.columns.str.replace(' ','_')
```

 # Feature engineering

```python
labels = ['Small', 'Medium', 'Large', 'Very Large']
mydata['AREA_TYPE'] = pd.qcut(mydata['AREA'], q=4, labels=labels)
```

```python

mydata['Zone_Raw'] = mydata['ADDRESS'].str.split(',').str[-2].str.strip()
```

```python
zone_map = {
    'Noida': 'Noida',
    'Gurgaon': 'Gurgaon',
    'South Delhi': 'South Delhi',
    'Central Delhi': 'Central Delhi',
    'North Delhi': 'North Delhi',
    
}

mydata['Zone'] = mydata['Zone_Raw'].map(zone_map)
mydata['Zone'] = mydata['Zone'].fillna('Other')  # Handle unmapped addresses
```

```python
mydata[['ADDRESS', 'Zone_Raw', 'Zone']].head(10)
mydata['Zone'].value_counts()
```

```python
mydata.head()
```

```python
mydata = mydata.drop(columns=['Zone'], errors='ignore')
```
